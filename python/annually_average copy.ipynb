```python
import pandas as pd

# Create a DataFrame with dates and values
df = pd.DataFrame({'Data': [  '2022-12-31', 
                              '2022-09-30', 
                              '2022-06-30', 
                              '2022-03-31',
                              '2021-12-31', 
                              '2021-09-30', 
                              '2021-06-30', 
                              '2021-03-31',
                              '2020-12-31', 
                              '2020-09-30', 
                              '2020-06-30', 
                              '2020-03-31',],                   
                   'Receita': [ 10, 15, 20, 25, 
                                30, 35, 40, 45, 
                                50, 55, 60, 65]})
df
```

```python
# Create a DataFrame with dates and values
df_2 = pd.DataFrame({'Data': [  '2022-12-31',
                                '2021-12-31', 
                                '2020-12-31',
                            ],                   
                   'Receita':[  (10+15+20+25)/4, 
                                (30+35+40+45)/4, 
                                (50+55+60+65)/4
                            ]
                    })
df_2
```

```python


# Convert the 'Data' column to datetime format
df['Data'] = pd.to_datetime(df['Data'])

 # Create a new column 'Quarter' to store the quarter corresponding to each date
df['Year'] = df['Data'].dt.year

# Group the data by quarter and calculate the mean of each group
medias_trimestrais = df.groupby('Year')['Receita'].mean()

# Replace the original DataFrame data with the quarterly averages
df = pd.DataFrame({
    'Year': medias_trimestrais.index,
    'Receita': medias_trimestrais.values
})


df = df.sort_values('Year', ascending=0)
df


```
