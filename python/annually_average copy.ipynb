```python
# Create a DataFrame with dates and values
df = pd.DataFrame({'Data': [ '2022-12-31', 
                             '2022-09-30', 
                             '2022-06-30', 
                             '2021-03-31',
                             '2021-12-31', 
                             '2021-09-30', 
                             '2021-06-30', 
                             '2021-03-31',
                             '2020-12-31', 
                             '2020-09-30', 
                             '2020-06-30', 
                             '2020-03-31',],                   
                   'Receita': [ 10, 15, 20, 25, 
                                30, 35, 40, 45, 
                                50, 55, 60, 65]})

# Convert the 'Data' column to datetime format
df['Data'] = pd.to_datetime(df['Data'])

# Create a new column 'quarter' to store the quarter corresponding to each date
df['quarter'] = df['Data'].dt.quarter

# Group the data by quarter and calculate the mean of each group
quarterly_avg = df.groupby('quarter')['Receita'].mean()

# Imprime as m√©dias trimestrais calculadas automaticamente
print('\nQuarterly averages calculated automatically:')
print(quarterly_avg)

df
```

```python


#  Converte a coluna 'Data' para o formato datetime
df['Data'] = pd.to_datetime(df['Data'])

# Cria uma nova coluna 'Ano' para armazenar o ano correspondente a cada data
df['Ano'] = df['Data'].dt.year

# Agrupa os dados por ano e calcula a m√©dia de cada grupo
media_anual = df.groupby('Ano')['Receita'].mean()

# Imprime a m√©dia anual
print(media_anual)

```
